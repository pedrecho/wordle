```python
letters = []
for i in range(32):
  letters.append(chr(i + ord('А')))
print(letters)
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://ozhegov.slovaronline.com/articles/'
dictionary = []
```

```python
for i in letters:
  k = 1
  link = url + i + '/page-' + str(k)
  page = requests.get(link)
  status = page.status_code
  print(link, status)
  while status == 200:
    soup = BeautifulSoup(page.text, "html.parser")
    words = soup.find('div', class_='articles-link-list').findAll('a')
    print(len(words))

    for l in words:
      if len(l.text) == 5:
        dictionary.append(l.text.lower())

    k += 1
    link = url + i + '/page-' + str(k)
    page = requests.get(link)
    status = page.status_code
    print(link, status)
```

```python
print(len(dictionary))
print(dictionary)
```

```python
open('dictionary.txt', 'w').write('\n'.join(dictionary))
```
